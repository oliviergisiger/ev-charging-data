## Example Query

The basic idea is to separate "dynamic" data from static data. 
The only data that is dynamic is the ``EvseStatus`` column, containing information about current status of the EV Charging Station. 

To query data, follow this pattern: 

1. define ``start`` and ``end``
2. load data in memory
3. filter on ``df_stations``, e.g., for ``Accessibility == "Free publicly accessible"``
4. merge ``df_stations`` and ``df``





```python
# 1.:  
start, end = date(2025, 7, 16), date(2025, 7, 18)

# 2.: 
df_stations = get_station_data(start)
df = get_time_series_data(date(2025, 7, 16), date(2025, 7, 20))

# 3.:
df_stations_filer = df_stations[df_stations.Accessibility == 'Free publicly accessible']
print(f'Rows, columns in df_stations_filter: {df_stations_filer.shape}')

# 4.: 
df_full = pd.merge(df_stations_filer, df, on='_id')
df_full.head()
```
